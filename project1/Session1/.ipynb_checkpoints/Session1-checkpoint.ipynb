## 1.1.3 Формирование итогового массива для хранения данных

<span style="color:blue">Проанализируем тематику обращений и целевые значения кассификатора</span>.
<span style="color:blue">Видно, что в тульском датасете (прянике) есть ненужные обращения, исправим это в соответсвии с датасетом классификатора</span>.

```python
target = set(pd.unique(Tula_classif['Тематика классификатора'].apply(lambda x: x.lower().strip())))
app = set(pd.unique(Tula_dataset['Наименование категории'].apply(lambda x: x.lower().strip())))

map_target = {}
key = app - target

for i in app:
    if 'вопросы по услугам мфц' in i:
        map_target[i] = 'вопросы по услугам мфц, портала госуслуги71'
    elif i in key:
        map_target[i] = 'Другое'.lower()
    else:
        map_target[i] = i

map_target
```

```python
Tula_classif['Тематика классификатора'] = Tula_classif['Тематика классификатора'].apply(lambda x: x.lower().strip())
Tula_dataset['Наименование категории'] = Tula_dataset['Наименование категории'].apply(lambda x: x.lower().strip())
Tula_dataset['Наименование категории'] = Tula_dataset['Наименование категории'].map(map_target)
```

```python


import spacy  # для лемматизации и контрольной проверки стоп-слов
from nltk.stem.snowball import SnowballStemmer  #  для стемминга
import re  # для очистки текста

stemmer = SnowballStemmer(language='russian')
nlp = spacy.load('ru_core_news_lg')

```

<span style="color:blue">Примеры, как будем работать с текстом</span>.

```python
# Нормализуем текст: произведем лемматизацию  стемминг
for doc in nlp('Привет мир. А так проводятся работы на улице Ленина 156 дом 34'):
    print(doc, doc.lemma_, stemmer.stem(doc.lemma_), sep='\t')
```

```python
# При необходимости можно извлекать сущности из текста (в данном случае принято решение не использовать)
for entity in nlp('А так Гарри Поттер проводит работы на улице Ленина 156 д. 34').ents:
    print(entity.text, entity.label_, sep='\t')
```

```python
# Пример использования регулярных выражений для полной очистки теста

words_pattern = '[а-я]+'

sen = '''бордюрами трамвайных путей по ул Лрайона$$22291$$в работа \r\n 2020ева до ул. Куйбышева.  
Электронная приёмная администрации Екатеринбурга: Екатеринбург.рф  Максимальный репост! 😈'''

re.findall(words_pattern, sen, flags=re.IGNORECASE)
```

<span style="color:blue">Функция обработки тектсовых данных:</span>.

```python
def preproccess(text):
    
    # Определим реугулярное выражение для очистки текста и выполняем токенизацию
    words_pattern = '[а-я]+'
    tokens = re.findall(words_pattern, text, flags=re.IGNORECASE)
    
    # Удалим все короткие слова, которые предположительно являются стоп-словами
    tokens = [token for token in tokens if len(token) >= 3]
    
    # Определим список конечных токенов
    tokens_list = list()
    
    # Выполним лемматизацию, стемминг (таким образом исправляя ошибки в обращениях) и
    # дополнительную проверку на стоп-слова
    for doc in nlp(" ".join(tokens)):
        if not doc.is_stop:
            tokens_list.append(stemmer.stem(doc.lemma_))
    
    return " ".join(tokens_list)
```

<span style="color:blue">Выполним обработку исключений при необходимости:</span>.

```python
%%time


try:
    df.text = df.text.apply(preproccess)
except Exception as exp:
    print(exp)
```
