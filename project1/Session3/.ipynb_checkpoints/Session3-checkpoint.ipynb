```python
df.category.replace({'текущее содержание многоквартирного дома':'жкх'}, inplace=True)
df.category.replace({'СПАМ':'другое'}, inplace=True)
df.category.replace({'безопасность':'другое'}, inplace=True)
```

# Кластеризация обращения граждан по категориям

<span style="color:blue">**Для задачи кластеризации будут испоьзованы седующие алгоритмы:**</span>.

<span style="color:blue">**1) KMeans - один из лучших простых и интерпретируемых алгоритмов кластеризации. В данном случае минимизируется отклонение точек от центроид кластеров**</span>.
        
<span style="color:blue">**2) MiniBatchKMeans - модификация KMeans, менее ресурсоемка**</span>.

<span style="color:blue">**3) SpectralClustering - рассматривает набор данных как граф и идентифицирует узлы графа на основе ребер. Данный метод очень гибкий**</span>.

<span style="color:blue">**Так как задачи кластеризации вычислительно сложны и ресурсоемки, будем использовать алгоритмы понижения размерности. Обучим модель понижения размерности PCA для преобразования признаков из исходной размерности в меньшее. Это поможет повысить скорость обучения моделей почти не изменяя точность**</span>.

<span style="color:blue">**Используем метод главных компонент, чтобы убрать шум в данных. Из графика ниже видно, что 90% иформации содержат первые 300 главных компонент, их и оставим, тем самым повысив произвдительность модели**</span>.

```python
pca = PCA(random_state=2022)
pca.fit(SelectKBest(chi2, k=500).fit_transform(TfidfVectorizer(lowercase=True, analyzer='word', max_features=1_000).
                                               fit_transform(X_train_category), y_train_category).todense())
skplt.decomposition.plot_pca_component_variance(pca, target_explained_variance=0.9)
```

<span style="color:blue">**Визуализируем категории обращения граждан в 2-D проекции**</span>.

```python
skplt.decomposition.plot_pca_2d_projection(pca, SelectKBest(chi2, k=500).
                                           fit_transform(TfidfVectorizer(lowercase=True, analyzer='word', 
                                                                         max_features=1_000).fit_transform(
                                               X_train_category), y_train_category).todense(), 
                                           y_train_category, figsize=(10, 7))
```

<span style="color:blue">**Создадим функцию для создания пайплайна и функцию вывода метрик кластеризации: 'homogeneity', 'completeness', 'measure', 'adjusted_rand_score'**</span>.

```python
def make_pipe_cluster(model, param: dict, n_comp: int):
    pipe = make_pipeline(
        TfidfVectorizer(lowercase=True, analyzer='word', max_features=1_000),
        SelectKBest(chi2, k=500),
        TruncatedSVD(n_components=n_comp),
        model(**param)
    )
    
    return pipe
```

```python
def print_metrics(model, x, y):
    # Первая метрика
    print('homogeneity_completeness_v_measure:', 
          metrics.homogeneity_completeness_v_measure(model.predict(x), y))
    # Вторая метрика
    print('adjusted_rand_score:', 
          metrics.adjusted_rand_score(model.predict(x), y))
```

<span style="color:blue">**Обучим первую модель KMeans & Выведем результат работы алгоритма, измерив ее с помощью соответствующих метрик**</span>.

```python
pipe = make_pipe_cluster(KMeans, 
                         param=dict(n_clusters=len(pd.unique(y_train_category)), random_state=True), n_comp=300)
pipe.fit(X_train_category, y_train_category)
# Тест
print_metrics(pipe, X_test_category, y_test_category)
```

# Классификация обращений по ведомствам

<span style="color:blue">**Создадим пайплайн классификации на логистической регресиии, посчитаем метрики 'precision', 'recall', 'f1-score', 'accuracy' и визуализируем матрицу неточностей и ROC-AUC. Произведем кросс-валидацию и поиск по сетке**</span>.

```python
def make_pipe_classif(model, n_comp: int):
    pipe = make_pipeline(
        TfidfVectorizer(lowercase=True, analyzer='word', max_features=1_000),
        SelectKBest(chi2, k=500),
        TruncatedSVD(n_components=n_comp),
        model()
    )
    
    return pipe
```

```python
logit_model = make_pipe_classif(LogisticRegression, n_comp=300)

params = {
    'logisticregression__penalty': ['l1', 'l2'],
    'logisticregression__C': [0.05, 0.1, 1]
}

grid_search = GridSearchCV(logit_model, params, cv=5, scoring='accuracy')
grid_search.fit(X_train_ved, y_train_ved)

print(metrics.classification_report(grid_search.predict(X_train_ved), y_train_ved))
print(metrics.classification_report(grid_search.predict(X_test_ved), y_test_ved))
```

```python
print(grid_search.best_params_, grid_search.best_score_)
```

```python
skplt.metrics.plot_confusion_matrix(grid_search.predict(X_test_ved), y_test_ved, normalize=True, figsize=(7, 7))
```

```python
skplt.metrics.plot_roc(y_test_ved, grid_search.predict_proba(X_test_ved), figsize=(10, 7))
```

```python
# Применем градиентный бустинг, и выведем метрики на тренировочном наборе и на тестовом
```

```python
# Создадим одну функцию, чтобы не переписывать
```

```python
def fit(model, params: dict, X_train_ved, X_test_ved, y_train_ved, y_test_ved):
    model = make_pipe_classif(model, n_comp=300)
    grid_search = GridSearchCV(model, params, cv=5, scoring='accuracy')
    grid_search.fit(X_train_ved, y_train_ved)
    print(metrics.classification_report(grid_search.predict(X_train_ved), y_train_ved))
    print(metrics.classification_report(grid_search.predict(X_test_ved), y_test_ved))
    print(grid_search.best_params_, grid_search.best_score_)
    skplt.metrics.plot_confusion_matrix(grid_search.predict(X_test_ved), y_test_ved, normalize=True, figsize=(7, 7));
    skplt.metrics.plot_roc(y_test_ved, grid_search.predict_proba(X_test_ved), figsize=(10, 7));
    return grid_search
```

<span style="color:blue">**Создадим пайплайн классификации на K-ближайших соседях, посчитаем метрики 'precision', 'recall', 'f1-score', 'accuracy' и визуализируем матрицу неточностей и ROC-AUC. Произведем кросс-валидацию и поиск по сетке**</span>.

```python
params = {
    'kneighborsclassifier__n_neighbors': [3, 6]
}
fit(KNeighborsClassifier, params, X_train_ved, X_test_ved, y_train_ved, y_test_ved)
```

<span style="color:blue">**Исходя из результатов и метрик " precision, recall, f1-score" лучше себя показал случайный лес, если судить по метрике "accuracy" - то логистическая регрессия**</span>.
