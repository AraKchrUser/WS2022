```python
for col in data.columns:
    if sum(data[col].isna())/len(data) > 0.4:
        print(col)
        data.drop([col], axis=1, inplace=True)
```

### 3. Описание набора данных

<!-- Опишем атрибуты набора данных. Для этого вычислим их статистики, в частности среднюю и медиану - для числовых, и моду - для категориальных -->

```python
aver = []
nanv = []
for col in data[1:]:
    if data[col].dtype == float:
        nanv.append(sum(data[col].isna())/len(data[col]))
        aver.append(data[col].dropna().sum()/len(data[col].dropna()))
```

```python
data.columns[1:]
```

```python
decription = ['Continent of the geographical location', 'Geographical location', 
              'Date of observation', 'Total confirmed cases of COVID-19. Counts can include probable cases, where reported',
             'New confirmed cases of COVID-19. Counts can include probable cases, where reported. In rare cases where our source reports a negative daily change due to a data correction, we set this metric to NA',
             'New confirmed cases of COVID-19 (7-day smoothed). Counts can include probable cases, where reported',
              'Total deaths attributed to COVID-19. Counts can include probable deaths, where reported', 
              'New deaths attributed to COVID-19. Counts can include probable deaths, where reported. In rare cases where our source reports a negative daily change due to a data correction, we set this metric to NA'] + (30 - 8) * ['Number of COVID-19 patients in hospital on a given day']
```

```python
for col in data.columns[1:]:
    if data[col].dtype == object:
        nanv.insert(0, sum(data[col].isna())/len(data[col]))
        aver.insert(0, data[col].value_counts().index[0])
```

### 4. Формирование дополнительных атрибутов

<!-- Рассмотрим случаи заражения в каждой стране в одно время -->

```python
data.groupby(['date', 'location'])['total_cases'].sum()
```

<!-- Посчитаем общее количество заражений в одно время  -->

```python
date_total_cases = data.groupby('date')['total_cases'].sum()
```

```python
date_total_cases
```

<!-- Рассчитаем коэффициент заражения  -->

```python
R_t = [0 for i in range(len(date_total_cases))]

for i in range(len(date_total_cases) - 1, 6, -1):
    last = 0
    previos = 0
    for j in range(4):
        last += date_total_cases[i - j]
        previos += date_total_cases[i - 4 - j]
    if previos != 0:
        R_t[i] = last / previos
```

```python
print(R_t[:50])
```

<!-- Отобразим матрицу корреляций для выявления зависимости коэффициента заражения и числа зараженных -->

```python
sns.heatmap(pd.DataFrame({'total_cases': date_total_cases.values, 'R_t':R_t}).corr())
```

```python
# Заполним пропуски, выполним понижение размерности, выберим алгоритм, произведем поиск по сетке, выведем метрики
```

```python
data = pd.read_csv('session1_data.csv', index_col=0)
```

<!-- Разметим данные - найдем среднее и смещение дисперсии и определим уровни опасности -->

```python
# Перемещаем данные
data = data.sample(frac=1)
```

<!-- Распределения данных далеки от нормальных, поэтому не обосновано использовать классические статистики, такие как
среднее или медиана, поэтому используем просто - моду -->

```python

imputer = impute.SimpleImputer(missing_values=np.nan, strategy='most_frequent')
data = imputer.fit_transform(data)
```

```python
data = pd.DataFrame(data)
```

<!-- Закодируем категориальные переменные -->

```python
le_continent = preprocessing.LabelEncoder()
data[0] = le_continent.fit_transform(data[0])
```

```python
le_continent.classes_
```

```python
le_country = preprocessing.LabelEncoder()
data[1] = le_country.fit_transform(data[1])
```

```python
len(le_country.classes_)
```

<!-- Преведем даты в дни от начала первых заражений -->

```python
data[2] = pd.to_datetime(data[2])
v = min(data[2])  # Timestamp('2020-01-01 00:00:00')
data[2] = data[2].apply(lambda x: (x - v).days)
```

<!-- Создадим переменную - уровень опасности и рассчитам ее исходя из среднего числа заражений и ее отклонения -->

```python
mean, std = pd.read_csv('session1_data.csv')['total_cases'].mean(), pd.read_csv('session1_data.csv')['total_cases'].std()
```

```python
mean, std
```

```python
def prepr(x):
    if x <= mean - std//2:
        return 0
    if x > mean + std//2:
        return 2
    return 1

data['y'] = pd.read_csv('session1_data.csv')['total_cases'].apply(prepr)
```

```python
y = pd.read_csv('session1_data.csv')['total_cases'].apply(prepr)
```

```python
data.head()
```

```python
data.y.value_counts()
```

<!-- Рассмотрим возможность разделения данных в пространстве более низкой размерности -->

```python
skplt.decomposition.plot_pca_2d_projection(decomposition.PCA().fit(data.iloc[:, :-1]),
                                                                   data.iloc[:, :-1], data.iloc[:, -1])
```

<!-- Нормализуем данные для устранения проблем с вычислением метрик (проклятие размерности) -->

```python
data = data.sample(n=len(data)//3)
sc = preprocessing.StandardScaler()
y = data.iloc[:, -1]
data = pd.DataFrame(sc.fit_transform(data.iloc[:, :-1]))
```

<!-- Выберем метод - K-средних, как наиболее простой и интерпретируемый алгоритм класстеризации. 
Также данный алгоритм очень производительный -->

```python

k_means = cluster.KMeans(n_clusters=3).fit(data)
```

<!-- Выполним проверку согласованности в кластерах данных -->

```python
skplt.metrics.plot_silhouette(data, k_means.predict(data))
```

<!-- Как видно из графика выше - уровень согласованности модели невысокий. -->

<!-- Выведем метрики качества модели -->

```python

metrics.homogeneity_completeness_v_measure(y, k_means.predict(data))
```

<!-- Рассмоторим другие алгоритмы -->

```python
# По-батчевый k-means позволит решить проблему переполнения памяти
k_means_mb = cluster.MiniBatchKMeans(n_clusters=3).fit(data)
skplt.metrics.plot_silhouette(data, k_means_mb.predict(data))
```

```python
metrics.homogeneity_completeness_v_measure(y, k_means_mb.predict(data))
```

<!-- Спектральная бикластеризация - очень производительный алгоритм. Работает с базовым предположением, что данные имеют структуру шахматной доски -->

```python
meanshift = cluster.SpectralBiclustering(n_clusters=3).fit(data)
```

```python
metrics.homogeneity_completeness_v_measure(y, meanshift.row_labels_)
```

```python
skplt.metrics.plot_silhouette(data, meanshift.row_labels_)
```
